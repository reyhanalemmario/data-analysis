# Relasi Suhu, Konsentrasi CO2 dan Dust Ppm di Atmosfer
Dokumen ini menampilkan visualisasi data parameter dalam bentuk deret waktu yang bertujuan untuk mendapatkan analisis dan jawaban atas tampilan yang dihasilkan. Data yang digunakan didapat dari Dosen yang berisikan nilai dari parameter yang akan dianalisis dari 400.000 tahun sebelum sekarang.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import numpy as np
```

```python
data = pd.read_csv('ppf.csv')
data2 = data.drop(columns = 'dust')
data2.dropna()
data2.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))

sns.lineplot(x='year', y='temp', data=data2, ax=ax, color = 'blue')
plt.ylabel('Suhu', fontsize = 16, color = 'blue')
plt.xlabel('Tahun Sebelum Sekarang', fontsize = 18)
ax2 = ax.twinx()
sns.lineplot(x='year', y='co2', data=data2, ax=ax2, color='r')
plt.ylabel('Konsentrasi CO2',fontsize = 18, color = 'r')
plt.title("Grafik Suhu dan Konsentrasi CO2 pada Waktu Lampau di Antartika", fontsize = 20)
plt.xlabel('Tahun Sebelum Sekarang', fontsize = 16)
#plt.savefig('Grafik Suhu dan Konsentrasi CO2 pada Waktu Lampau di Antartika.png')
plt.show()
```

Terlihat bahwa fluktuasi konsentrasi CO2 mempengaruhi Fluktuasi suhu yang hubungan korelasinya akan dicari pada line selanjutnya.

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='co2', y='temp', data=data2, color = 'g')
sns.regplot(x='co2', y='temp', data=data2, color='g')
plt.xlabel('Konsentrasi CO2', fontsize = 18)
plt.ylabel('Suhu', fontsize = 18)
plt.title('Korelasi antara Konsentrasi CO2 dengan Suhu di Antartika', fontsize = 20)
plt.savefig('Korelasi antara Konsentrasi CO2 dengan Suhu di Antartika.png')
plt.show()
```

```python
data3 = data.drop(columns = 'co2')
data3.dropna()
data3.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))

sns.lineplot(x='year', y='temp', data=data3, ax=ax, color = 'blue')
plt.ylabel('Suhu', fontsize = 16, color = 'blue')
plt.xlabel('Tahun Sebelum Sekarang', fontsize = 18)
ax.set_ylim([-10,16])
ax2 = ax.twinx()
sns.lineplot(x='year', y='dust', data=data3, ax=ax2, color='r')
plt.ylabel('Dust PPM',fontsize = 18, color = 'r')
plt.title("Grafik Suhu dan Dust PPM pada Waktu Lampau di Antartika", fontsize = 20)
plt.xlabel('Tahun Sebelum Sekarang', fontsize = 16)
ax2.set_ylim([-1,1.8])
#plt.savefig('Grafik Suhu dan Dust PPM pada Waktu Lampau di Antartika.png')
plt.show()
```

Terlihat bahwa terdapat hubungan yang terbalik antara Suhu dan Dust Ppm, hubungannya terlihat tidak linear karena seperti mempunyai batas tertentu untuk parameter Suhu.

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='dust', y='temp', data=data3, color = 'g')
sns.regplot(x='dust', y='temp', data=data3, color='g', logx = True)
plt.xlabel('Dust PPM', fontsize = 18)
plt.ylabel('Suhu', fontsize = 18)
plt.title('Korelasi antara Dust PPM dengan Suhu di Antartika', fontsize = 20)
plt.savefig('Korelasi antara Dust PPM dengan Suhu di Antartika.png')
plt.show()
```

Setelah ditampilkan hubungannya, maka jelas terlihat memang ada hubungan yang berkebalikan antara Suhu dan Dust Ppm, namun untuk Dust Ppm yang nilainya semakin besar, nilai Suhu relatif tidak berubah.
